```python
#"Which channel should we use?"
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_absolute_error, r2_score, accuracy_score, classification_report
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')
```

```python
#Encode
# ================================================
# ENCODING ALL CATEGORICAL VARIABLES
# ================================================

# Create a copy for encoding
df_encoded = data.copy()

# Clean 'Acquisition_Cost' column: remove '$' and convert to float
df_encoded['Acquisition_Cost'] = df_encoded['Acquisition_Cost'].replace({r'[$,]': ''}, regex=True).astype(float)

# 1. One-Hot Encoding for low-cardinality features
print("1. One-Hot Encoding...")

# Channel_Used (this is crucial for channel recommendation)
channel_dummies = pd.get_dummies(df_encoded['Channel_Used'], prefix='Channel')
df_encoded = pd.concat([df_encoded, channel_dummies], axis=1)
print(f"   - Created {channel_dummies.shape[1]} channel dummies")

# Campaign_Goal
goal_dummies = pd.get_dummies(df_encoded['Campaign_Goal'], prefix='Goal')
df_encoded = pd.concat([df_encoded, goal_dummies], axis=1)
print(f"   - Created {goal_dummies.shape[1]} goal dummies")

# Location (if reasonable number of locations)
if df_encoded['Location'].nunique() < 20:
    location_dummies = pd.get_dummies(df_encoded['Location'], prefix='Location')
    df_encoded = pd.concat([df_encoded, location_dummies], axis=1)
    print(f"   - Created {location_dummies.shape[1]} location dummies")

# Language
language_dummies = pd.get_dummies(df_encoded['Language'], prefix='Language')
df_encoded = pd.concat([df_encoded, language_dummies], axis=1)
print(f"   - Created {language_dummies.shape[1]} language dummies")

# Customer_Segment
segment_dummies = pd.get_dummies(df_encoded['Customer_Segment'], prefix='Segment')
df_encoded = pd.concat([df_encoded, segment_dummies], axis=1)
print(f"   - Created {segment_dummies.shape[1]} segment dummies")

# 2. Label Encoding for high-cardinality features
print("\n2. Label Encoding...")

# Target_Audience (might have many unique values)
le_target = LabelEncoder()
df_encoded['Target_Audience_Encoded'] = le_target.fit_transform(df_encoded['Target_Audience'])
print("   - Encoded Target_Audience")

# Company (many unique values)
le_company = LabelEncoder()
df_encoded['Company_Encoded'] = le_company.fit_transform(df_encoded['Company'])
print("   - Encoded Company")

# 3. Extract features from Date
print("\n3. Extracting date features...")
df_encoded['Date'] = pd.to_datetime(df_encoded['Date'])
df_encoded['Month'] = df_encoded['Date'].dt.month
df_encoded['Quarter'] = df_encoded['Date'].dt.quarter
df_encoded['Day_of_Week'] = df_encoded['Date'].dt.dayofweek
df_encoded['Is_Weekend'] = df_encoded['Day_of_Week'].isin([5, 6]).astype(int)
df_encoded['Day_of_Month'] = df_encoded['Date'].dt.day
df_encoded['Week_of_Year'] = df_encoded['Date'].dt.isocalendar().week
print("   - Created Month, Quarter, Day_of_Week, Is_Weekend, Day_of_Month, Week_of_Year")

# 4. Create engagement metrics
print("\n4. Creating derived metrics...")
df_encoded['CTR'] = (df_encoded['Clicks'] / df_encoded['Impressions'] * 100).round(2)
df_encoded['CPC'] = (df_encoded['Acquisition_Cost'] / df_encoded['Clicks']).round(2)
df_encoded['Cost_per_Impression'] = (df_encoded['Acquisition_Cost'] / df_encoded['Impressions'] * 1000).round(2)
df_encoded['Engagement_Rate'] = (df_encoded['Engagement_Score'] / df_encoded['Impressions'] * 100).round(4)
print("   - Created CTR, CPC, Cost_per_Impression, Engagement_Rate")

print(f"\n✅ Encoding complete! New shape: {df_encoded.shape}")
```

```python
# ================================================
# DEFINE FEATURES FOR EACH PLANNING QUESTION
# ================================================

# Base numeric features
numeric_features = ['Duration', 'Acquisition_Cost', 'Clicks', 'Impressions',
                    'Engagement_Score', 'Conversion_Rate']

# Derived metrics
derived_features = ['CTR', 'CPC', 'Cost_per_Impression', 'Engagement_Rate']

# Date features
date_features = ['Month', 'Quarter', 'Day_of_Week', 'Is_Weekend', 'Day_of_Month', 'Week_of_Year']

# Get all dummy columns
channel_cols = [col for col in df_encoded.columns if col.startswith('Channel_')]
goal_cols = [col for col in df_encoded.columns if col.startswith('Goal_')]
location_cols = [col for col in df_encoded.columns if col.startswith('Location_')]
language_cols = [col for col in df_encoded.columns if col.startswith('Language_')]
segment_cols = [col for col in df_encoded.columns if col.startswith('Segment_')]

# ================================================
# FEATURE SET FOR EACH PLANNING QUESTION
# ================================================

# 1. CHANNEL RECOMMENDATION FEATURES
channel_features = [
    # Campaign characteristics
    'Target_Audience_Encoded', 'Company_Encoded',
    'Duration', 'Acquisition_Cost',
    # Performance metrics of similar campaigns
    'Conversion_Rate', 'CTR', 'Engagement_Rate',
    # Temporal
    'Month', 'Quarter', 'Day_of_Week', 'Is_Weekend'
] + goal_cols + location_cols + language_cols + segment_cols

print(f"Channel Recommendation Features: {len(channel_features)}")

# 2. DURATION OPTIMIZATION FEATURES
duration_features = [
    # Channel and segment
    'Channel_Used', 'Customer_Segment', 'Target_Audience_Encoded',
    # Campaign goals
    'Campaign_Goal', 'Acquisition_Cost',
    # Performance indicators
    'Clicks', 'Impressions', 'Engagement_Score', 'Conversion_Rate',
    # Temporal
    'Month', 'Is_Weekend'
]
print(f"Duration Optimization Features: {len(duration_features)}")

# 3. BUDGET PREDICTION FEATURES
budget_features = [
    # Campaign setup
    'Channel_Used', 'Campaign_Goal', 'Customer_Segment', 'Location', 'Language',
    'Target_Audience_Encoded', 'Duration',
    # Target metrics
    'Clicks', 'Impressions', 'Conversion_Rate',
    # Temporal
    'Month', 'Quarter', 'Day_of_Week'
]
print(f"Budget Prediction Features: {len(budget_features)}")

# 4. TIMING OPTIMIZATION FEATURES
timing_features = [
    # All date features
    'Month', 'Quarter', 'Day_of_Week', 'Is_Weekend', 'Day_of_Month', 'Week_of_Year',
    # Campaign context
    'Channel_Used', 'Campaign_Goal', 'Customer_Segment', 'Location',
    # Performance
    'Conversion_Rate', 'Engagement_Score'
]
print(f"Timing Optimization Features: {len(timing_features)}")
```

## Channel Recommendation Engine

```python
# ================================================
# PROBLEM 1: "Which channel should we use?"
# ================================================

print("\n" + "="*60)
print("PROBLEM 1: Channel Recommendation Engine")
print("="*60)

# Prepare data for channel recommendation
# Target: Which channel performed best for similar campaigns

# Create a dataset with channel performance
channel_performance = df_encoded.groupby(['Channel_Used', 'Customer_Segment', 'Campaign_Goal']).agg({
    'ROI': 'mean',
    'Conversion_Rate': 'mean',
    'Engagement_Score': 'mean',
    'CTR': 'mean'
}).reset_index()

print("\nChannel Performance Summary:")
print(channel_performance.head(10))

# Method A: Rule-based recommendation
def recommend_channel_rule_based(segment, goal, location=None):
    """
    Simple rule-based recommendation using historical averages
    """
    # Filter historical data
    mask = (df_encoded['Customer_Segment'] == segment) & (df_encoded['Campaign_Goal'] == goal)
    if location:
        mask &= (df_encoded['Location'] == location)

    historical = df_encoded[mask]

    if len(historical) == 0:
        return "Insufficient data for this combination"

    # Calculate average ROI by channel
    channel_roi = historical.groupby('Channel_Used')['ROI'].mean().sort_values(ascending=False)

    return channel_roi.head(3)

# Test the rule-based recommender
test_segment = df_encoded['Customer_Segment'].iloc[0]
test_goal = df_encoded['Campaign_Goal'].iloc[0]
print(f"\nRule-based recommendation for {test_segment} - {test_goal}:")
print(recommend_channel_rule_based(test_segment, test_goal))

# Method B: ML-based recommendation (Multi-class classification)
print("\nMethod B: Building ML-based Channel Recommender...")

# Prepare features for ML model
X_channel = df_encoded[channel_features].copy()
y_channel = df_encoded['Channel_Used']

# Encode categorical features in X
for col in X_channel.select_dtypes(include=['object']).columns:
    X_channel[col] = LabelEncoder().fit_transform(X_channel[col].astype(str))

# Fill any missing values
X_channel = X_channel.fillna(0)

# Split data
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    X_channel, y_channel, test_size=0.2, random_state=42
)

# Train Random Forest Classifier
rf_channel = RandomForestClassifier(n_estimators=100, random_state=42)
rf_channel.fit(X_train_c, y_train_c)

# Evaluate
y_pred_c = rf_channel.predict(X_test_c)
accuracy = accuracy_score(y_test_c, y_pred_c)
print(f"Channel Recommendation Accuracy: {accuracy:.2%}")

# Feature importance for channel selection
channel_importance = pd.DataFrame({
    'Feature': X_channel.columns,
    'Importance': rf_channel.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nTop 10 factors influencing channel selection:")
print(channel_importance.head(10))

# Create recommendation function
def recommend_channel_ml(campaign_features):
    """
    ML-based channel recommendation
    """
    # Prepare input features (same as training)
    input_df = pd.DataFrame([campaign_features])

    # Encode and align columns
    for col in X_channel.columns:
        if col not in input_df.columns:
            input_df[col] = 0

    input_df = input_df[X_channel.columns]

    # Get predictions and probabilities
    prediction = rf_channel.predict(input_df)[0]
    probabilities = rf_channel.predict_proba(input_df)[0]

    # Get top 3 channels
    top_3_idx = probabilities.argsort()[-3:][::-1]
    top_3_channels = rf_channel.classes_[top_3_idx]
    top_3_probs = probabilities[top_3_idx]

    return {
        'top_channel': prediction,
        'top_3_channels': list(zip(top_3_channels, top_3_probs)),
        'confidence': max(probabilities)
    }

print("\n✅ Channel Recommendation Engine Ready!")
```

##Duration Optimization Model

```python
# ================================================
# PROBLEM 2: "How long should campaign run?"
# ================================================

print("\n" + "="*60)
print("PROBLEM 2: Duration Optimization Model")
print("="*60)

# Prepare data for duration prediction
# We want to find optimal duration based on ROI

# Convert 'Duration' to numeric by removing ' Days'
df_encoded['Duration'] = df_encoded['Duration'].str.replace(' Days', '').astype(int)

# Calculate ROI per day
df_encoded['ROI_per_Day'] = df_encoded['ROI'] / df_encoded['Duration']

# Find optimal duration for each channel-segment combination
optimal_duration = df_encoded.groupby(['Channel_Used', 'Customer_Segment']).apply(
    lambda x: x.loc[x['ROI'].idxmax(), 'Duration']
).reset_index()
optimal_duration.columns = ['Channel_Used', 'Customer_Segment', 'Optimal_Duration']

print("\nOptimal Duration by Channel and Segment:")
print(optimal_duration.head(10))

# Build ML model to predict optimal duration
# Features: Channel, Segment, Goal, Budget, etc.

# Prepare features
X_duration = df_encoded[['Channel_Used', 'Customer_Segment', 'Campaign_Goal',
                         'Acquisition_Cost', 'Location', 'Month']].copy()
y_duration = df_encoded['Duration']  # We'll predict duration that maximizes ROI

# Encode categorical features
le_dict = {}
for col in X_duration.select_dtypes(include=['object']).columns:
    le_dict[col] = LabelEncoder()
    X_duration[col] = le_dict[col].fit_transform(X_duration[col].astype(str))

# Train Random Forest Regressor
rf_duration = RandomForestRegressor(n_estimators=100, random_state=42)
rf_duration.fit(X_duration, y_duration)

# Evaluate
y_pred_d = rf_duration.predict(X_duration)
mae = mean_absolute_error(y_duration, y_pred_d)
r2 = r2_score(y_duration, y_pred_d)

print(f"\nDuration Prediction Performance:")
print(f"MAE: {mae:.2f} days")
print(f"R² Score: {r2:.3f}")

# Feature importance for duration
duration_importance = pd.DataFrame({
    'Feature': X_duration.columns,
    'Importance': rf_duration.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nFactors affecting optimal duration:")
print(duration_importance)

# Create duration recommendation function
def recommend_duration(channel, segment, goal, budget, location):
    """
    Recommend optimal campaign duration
    """
    # Prepare input
    input_data = {
        'Channel_Used': channel,
        'Customer_Segment': segment,
        'Campaign_Goal': goal,
        'Acquisition_Cost': budget,
        'Location': location,
        'Month': pd.Timestamp.now().month
    }

    # Convert to DataFrame and encode
    input_df = pd.DataFrame([input_data])
    for col, le in le_dict.items():
        if col in input_df.columns:
            try:
                input_df[col] = le.transform(input_df[col].astype(str))
            except:
                input_df[col] = -1  # Unknown category

    # Predict
    predicted_duration = rf_duration.predict(input_df)[0]

    # Get confidence interval (using standard deviation of similar campaigns)
    similar_campaigns = df_encoded[
        (df_encoded['Channel_Used'] == channel) &
        (df_encoded['Customer_Segment'] == segment)
    ]['Duration']

    if len(similar_campaigns) > 0:
        std_duration = similar_campaigns.std()
        confidence_interval = (predicted_duration - std_duration,
                              predicted_duration + std_duration)
    else:
        confidence_interval = (predicted_duration - 5, predicted_duration + 5)

    return {
        'recommended_duration': round(predicted_duration),
        'confidence_interval': (round(confidence_interval[0]), round(confidence_interval[1])),
        'min_days': max(1, round(predicted_duration - 7)),
        'max_days': round(predicted_duration + 7)
    }

# Test the function
test_rec = recommend_duration('Email', 'High Value', 'Conversion', 50000, 'North America')
print(f"\nTest Recommendation:")
print(f"Recommended duration: {test_rec['recommended_duration']} days")
print(f"Confidence interval: {test_rec['confidence_interval']} days")
print(f"Try range: {test_rec['min_days']} - {test_rec['max_days']} days")

print("\n✅ Duration Optimization Model Ready!")
```

##PROBLEM 3 - Budget Prediction Mode

```python
# ================================================
# PROBLEM 3: "What budget do we need?"
# ================================================

print("\n" + "="*60)
print("PROBLEM 3: Budget Prediction Model")
print("="*60)

# Prepare data for budget prediction
# We want to predict required budget to achieve target ROI/conversions

# Build features for budget prediction
X_budget = df_encoded[['Channel_Used', 'Campaign_Goal', 'Customer_Segment',
                       'Duration', 'Location', 'Language', 'Target_Audience_Encoded',
                       'Month', 'Quarter']].copy()
y_budget = df_encoded['Acquisition_Cost']  # Target: budget needed

# Encode categorical features
budget_encoders = {}
for col in X_budget.select_dtypes(include=['object']).columns:
    budget_encoders[col] = LabelEncoder()
    X_budget[col] = budget_encoders[col].fit_transform(X_budget[col].astype(str))

# Fill missing values
X_budget = X_budget.fillna(0)

# Split data
X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(
    X_budget, y_budget, test_size=0.2, random_state=42
)

# Train XGBoost Regressor (usually better for budget prediction)
xgb_budget = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
xgb_budget.fit(X_train_b, y_train_b)

# Evaluate
y_pred_b = xgb_budget.predict(X_test_b)
mae_b = mean_absolute_error(y_test_b, y_pred_b)
r2_b = r2_score(y_test_b, y_pred_b)

print(f"\nBudget Prediction Performance:")
print(f"MAE: ${mae_b:,.2f}")
print(f"R² Score: {r2_b:.3f}")

# Calculate accuracy within 20% range
within_20_percent = np.abs((y_test_b - y_pred_b) / y_test_b) <= 0.2
accuracy_20 = within_20_percent.mean()
print(f"Accuracy within 20%: {accuracy_20:.1%}")

# Feature importance
budget_importance = pd.DataFrame({
    'Feature': X_budget.columns,
    'Importance': xgb_budget.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nTop factors affecting budget:")
print(budget_importance.head(10))

# Create budget recommendation function
def recommend_budget(channel, goal, segment, duration, location, target_audience=None):
    """
    Recommend campaign budget based on historical data
    """
    # Prepare input
    input_data = {
        'Channel_Used': channel,
        'Campaign_Goal': goal,
        'Customer_Segment': segment,
        'Duration': duration,
        'Location': location,
        'Language': 'English',  # Default or make it a parameter
        'Target_Audience_Encoded': 0,  # Default
        'Month': pd.Timestamp.now().month,
        'Quarter': (pd.Timestamp.now().month - 1) // 3 + 1
    }

    # Convert to DataFrame
    input_df = pd.DataFrame([input_data])

    # Encode categorical features
    for col, le in budget_encoders.items():
        if col in input_df.columns:
            try:
                input_df[col] = le.transform(input_df[col].astype(str))
            except:
                input_df[col] = -1  # Unknown category

    # Ensure column order matches training
    input_df = input_df[X_budget.columns]

    # Predict
    predicted_budget = xgb_budget.predict(input_df)[0]

    # Get budget range based on similar campaigns
    similar = df_encoded[
        (df_encoded['Channel_Used'] == channel) &
        (df_encoded['Campaign_Goal'] == goal) &
        (df_encoded['Customer_Segment'] == segment)
    ]

    if len(similar) > 0:
        min_budget = similar['Acquisition_Cost'].quantile(0.25)
        max_budget = similar['Acquisition_Cost'].quantile(0.75)
    else:
        min_budget = predicted_budget * 0.7
        max_budget = predicted_budget * 1.3

    return {
        'recommended_budget': round(predicted_budget, -2),  # Round to nearest 100
        'budget_range': (round(min_budget, -2), round(max_budget, -2)),
        'confidence': 'High' if len(similar) > 10 else 'Medium' if len(similar) > 3 else 'Low'
    }

# Test budget recommendation
test_budget = recommend_budget(
    channel='Email',
    goal='Conversion',
    segment='High Value',
    duration=30,
    location='North America'
)

print(f"\nTest Budget Recommendation:")
print(f"Recommended budget: ${test_budget['recommended_budget']:,.0f}")
print(f"Budget range: ${test_budget['budget_range'][0]:,.0f} - ${test_budget['budget_range'][1]:,.0f}")
print(f"Confidence: {test_budget['confidence']}")

print("\n✅ Budget Prediction Model Ready!")
```
