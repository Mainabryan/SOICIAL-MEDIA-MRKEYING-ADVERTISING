```python
#"Which channel should we use?"
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_absolute_error, r2_score, accuracy_score, classification_report
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')
```

```python
#Encode
# ================================================
# ENCODING ALL CATEGORICAL VARIABLES
# ================================================

# Create a copy for encoding
df_encoded = data.copy()

# Clean 'Acquisition_Cost' column: remove '$' and convert to float
df_encoded['Acquisition_Cost'] = df_encoded['Acquisition_Cost'].replace({r'[$,]': ''}, regex=True).astype(float)

# 1. One-Hot Encoding for low-cardinality features
print("1. One-Hot Encoding...")

# Channel_Used (this is crucial for channel recommendation)
channel_dummies = pd.get_dummies(df_encoded['Channel_Used'], prefix='Channel')
df_encoded = pd.concat([df_encoded, channel_dummies], axis=1)
print(f"   - Created {channel_dummies.shape[1]} channel dummies")

# Campaign_Goal
goal_dummies = pd.get_dummies(df_encoded['Campaign_Goal'], prefix='Goal')
df_encoded = pd.concat([df_encoded, goal_dummies], axis=1)
print(f"   - Created {goal_dummies.shape[1]} goal dummies")

# Location (if reasonable number of locations)
if df_encoded['Location'].nunique() < 20:
    location_dummies = pd.get_dummies(df_encoded['Location'], prefix='Location')
    df_encoded = pd.concat([df_encoded, location_dummies], axis=1)
    print(f"   - Created {location_dummies.shape[1]} location dummies")

# Language
language_dummies = pd.get_dummies(df_encoded['Language'], prefix='Language')
df_encoded = pd.concat([df_encoded, language_dummies], axis=1)
print(f"   - Created {language_dummies.shape[1]} language dummies")

# Customer_Segment
segment_dummies = pd.get_dummies(df_encoded['Customer_Segment'], prefix='Segment')
df_encoded = pd.concat([df_encoded, segment_dummies], axis=1)
print(f"   - Created {segment_dummies.shape[1]} segment dummies")

# 2. Label Encoding for high-cardinality features
print("\n2. Label Encoding...")

# Target_Audience (might have many unique values)
le_target = LabelEncoder()
df_encoded['Target_Audience_Encoded'] = le_target.fit_transform(df_encoded['Target_Audience'])
print("   - Encoded Target_Audience")

# Company (many unique values)
le_company = LabelEncoder()
df_encoded['Company_Encoded'] = le_company.fit_transform(df_encoded['Company'])
print("   - Encoded Company")

# 3. Extract features from Date
print("\n3. Extracting date features...")
df_encoded['Date'] = pd.to_datetime(df_encoded['Date'])
df_encoded['Month'] = df_encoded['Date'].dt.month
df_encoded['Quarter'] = df_encoded['Date'].dt.quarter
df_encoded['Day_of_Week'] = df_encoded['Date'].dt.dayofweek
df_encoded['Is_Weekend'] = df_encoded['Day_of_Week'].isin([5, 6]).astype(int)
df_encoded['Day_of_Month'] = df_encoded['Date'].dt.day
df_encoded['Week_of_Year'] = df_encoded['Date'].dt.isocalendar().week
print("   - Created Month, Quarter, Day_of_Week, Is_Weekend, Day_of_Month, Week_of_Year")

# 4. Create engagement metrics
print("\n4. Creating derived metrics...")
df_encoded['CTR'] = (df_encoded['Clicks'] / df_encoded['Impressions'] * 100).round(2)
df_encoded['CPC'] = (df_encoded['Acquisition_Cost'] / df_encoded['Clicks']).round(2)
df_encoded['Cost_per_Impression'] = (df_encoded['Acquisition_Cost'] / df_encoded['Impressions'] * 1000).round(2)
df_encoded['Engagement_Rate'] = (df_encoded['Engagement_Score'] / df_encoded['Impressions'] * 100).round(4)
print("   - Created CTR, CPC, Cost_per_Impression, Engagement_Rate")

print(f"\n✅ Encoding complete! New shape: {df_encoded.shape}")
```
